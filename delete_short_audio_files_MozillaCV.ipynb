```python
import os
import librosa

def delete_short_audio_files(directory_path):
  """Deletes audio files shorter than 4 seconds in a directory.

  Args:
    directory_path: The path to the directory containing audio files.
  """
  # Get all audio files with the specified extension (default: wav)
  for filename in os.listdir(directory_path):
    if filename.endswith(".mp3"):  # Change extension if needed
      file_path = os.path.join(directory_path, filename)

      # Open the audio file
      duration = librosa.get_duration(filename=file_path)

      # Check if audio duration is less than 4 seconds
      if duration < 4:
        # Delete the file
        os.remove(file_path)
        print(f"Deleted: {filename}")
```

```python
delete_short_audio_files('data/speechCorp/cv-corpus-16.1-2023-12-06/nan-tw/clips/')
```
